# Introduction
This project aims to enhance the SpeechT5 model's capabilities for text-to-speech (TTS) tasks, with a focus on accurately handling technical terms and domain-specific language. Originally, the model was trained on general speech data, which might not effectively address the nuances of specialized vocabulary. Therefore, fine-tuning the model for targeted applications is essential.

## Dataset Description
For fine-tuning the SpeechT5 model, we utilized a dataset that combines general speech data in Hindi. This dataset serves to instruct the model on the correct pronunciation of sentences in the Hindi language.

We selected this dataset due to its diverse range of Hindi sentences and vocabulary, making it well-suited for enhancing the model's proficiency with specialized language.

The dataset is composed of text-to-speech (TTS) pairs, where each sample includes a text phrase and its corresponding audio file. The dataset contains:

- **Total Samples:** Approximately 42,2000
- **Duration:** Approximately 1000 hours of speech data
- **Text Length:** Ranges from single words to full sentences.

The text data encompasses both general speech and various tonalities. This diversity allows the model to learn not only general speech patterns but also specific vocabulary related to different domains.


## Data Split
The dataset was split into three subsets:

- **Training Set:** 90% of the data was used for training the model.
- **Test Set:** The remaining 10% was reserved for final model evaluation.

This split ensures that the model has sufficient data to learn while providing a reliable test set for unbiased performance evaluation.

## Model Performance
During the fine-tuning of the SpeechT5 model, we tracked several metrics to monitor the training process, including training loss, validation loss, and accuracy. These metrics help us understand the model's convergence and how well it is learning from the dataset.



```python
import numpy as np

data = """0.8562,0.3366,100,0.7391
0.7663,0.6731,200,0.6572
0.7168,1.0097,300,0.6354
0.6915,1.3462,400,0.6043
0.6651,1.6828,500,0.6000
0.6573,2.0194,600,0.5836
0.6347,2.3559,700,0.5818
0.6554,2.6925,800,0.5786"""

# Split the data and convert to NumPy array
lines = data.split('\n')
data_array = np.array([[float(value) for value in line.split(',')] for line in lines])

# Direct arrays for each column
epoch_values = data_array[:, 1]
training_loss_values = data_array[:, 0]
step_values = data_array[:, 2]
validation_loss_values = data_array[:, 3]

# Print the arrays
print("Epoch Values:", epoch_values)
print("Training Loss Values:", training_loss_values)
print("Step Values:", step_values)
print("Validation Loss Values:", validation_loss_values)
```

```python
import matplotlib.pyplot as plt

epochs = [0.3366, 0.6731, 1.0097, 1.3462, 1.6828, 2.0194, 2.3559, 2.6925]
training_loss = [0.8562, 0.7663, 0.7168, 0.6915, 0.6651, 0.6573, 0.6347, 0.6554]
validation_loss = [0.7391, 0.6572, 0.6354, 0.6043, 0.6 , 0.5836, 0.5818, 0.5786]

plt.plot(epochs, training_loss, 'b-', label='Training Loss')
plt.plot(epochs, validation_loss, 'r-', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

The graph above shows the training and validation loss over 8 epochs. As the model trains, the loss decreases but at the end the loss slightly increases, indicating that the model is learning but not properly and might not predict the speech from text more accurately. The validation loss also decreases, which suggests that the model generalizes well to unseen data.
